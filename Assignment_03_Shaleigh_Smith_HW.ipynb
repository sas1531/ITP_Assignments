### Q3. Create 2 python functions that: 
<br> </br>
#### A. Creates/*Returns* a new DataFrame that displays the number of missing values in every column. 
#### This new DataFrame has one new column named "Missing" with the sum of the missing values from the columns

### Notes 
***Function takes one argument: An input DataFrame***
<br> </br>
<br> </br>
** One Proposed method:**
1. Create a pandas series object from the sum of the nulls in every column (review class notes)
2. Convert the series object to a DataFrame and pass the list of column names from the input DataFrame as the index. (read Pandas Help online)
3. During the conversion of the series object to a DataFrame pass a columns argument to create a name for the new Output DataFrame's Column.

### Expected Output:
<img src='Missing_cols_expectedoutput.png'>

```python
# Answer 3A: CODE HERE; Please comment each part

def Missing_Cols(df):
    null = df.isnull().sum(axis=0)   #sum all of the NaN values in each column and return them as integers in a series
    d = pd.DataFrame(data = null)    #convert the series into a data frame
    d = d.rename(index = str, columns = {0:'Missing'})    #rename the NaN count to 'Missing'
    return d                                              #return the data frame 
    
Missing_Cols(NHL_Goalies)
```

#### B. Creates a new *column* called 'missing_values' in the input DF that sums the missing values in each row.

### Notes: 
***Function takes one argument: An input DataFrame***
<br> </br>
<br> </br>
Review the Column creation method in lecture and the axis arguments needed

### Expected Output:
<img src= 'missingrows_expectedoutput.png'>

```python
#Answer 3B here; Please comment each part

def Missing_Rows(df):
    null = df.isnull().sum(axis=1)   #sum all of the NaN values in each row and return them as integers in a series
    d = pd.DataFrame(data = null)    #convert the series into a data frame
    d = d.rename(columns = {0:'missing_values'})    #rename the NaN count to 'missing_values'
    df = pd.concat([df, d], sort = True, axis=1)    #concatenate the orignal dataframe and the altered dataframe
    return df                                       #return the data frame 
    
Missing_Rows(NHL_Goalies)
```

```python
#Answer 5 here: please comments (see below)

#5.1
#subset the injries by only looking at the null values

NHL_Goalies_NaN = (NHL_Goalies[NHL_Goalies.Injuries.isnull()])
NHL_Goalies_NaN[['Injuries']]
```

```python
#5.2
# merge the two data frames by inner merge in order to only get the intersection
# Take from both the right and left indices (index = true)

mergedDF = pd.merge(NHL_Goalies_NaN, GAA_, right_index=True, left_index=True, how = 'inner')
#mergedDF
#list(mergedDF.columns.values)
```

```python
#5.3
# isolate only specific columns (used x when there was a duplicate)

mergedDF = mergedDF[['First Name_x', 'Last Name_x', 'Team_x', 'Cntry', 'SV%_x', 'GA_x', 'GAA_x', 'TOI']]
mergedDF
```

```python
#5.4
# group by country
# use the dataframe agg function to calculate the mean, min, max of the viable colums

mergedDF.groupby('Cntry').agg({'SV%_x':['mean', 'min', 'max'], 
                               'GA_x':['mean', 'min', 'max'],
                               'GAA_x':['mean', 'min', 'max'],
                               'TOI':['mean', 'min', 'max'], })
```

```python
#5.5
# isolate the TOI values
# use the to_csv function to create the comma separated values file 

TOI = mergedDF.groupby('Cntry').agg({'TOI':['mean', 'min', 'max']})
TOI.to_csv()
```

## that subsets a dataframe by removing rows that are WITHIN the InterQuartile Region of a specified column

### Add a condition that returns an error if the Column Name chosen is not:

#hint: There are several pandas subfunctions that can be useful

```python
#Answer Question 6 here

import sys
from pandas.api.types import is_numeric_dtype

def remove(df, col):
    
    #check if the data type is numeric, if not return error
    if is_numeric_dtype(df[col]) == False:
        sys.exit('The data type is not numeric')
    
    #select for only numeric datatypes
    #if the column is not included in the list of datatypes, return error
    df_numeric = df.select_dtypes(include = ['float64', 'int64'])
    if col not in df_numeric:
        sys.exit('Column does not exist in the numeric columns')
        
    #Define IQR
    Q1 = df[col].quantile(0.25)  #find the first quartile
    Q3 = df[col].quantile(0.75)  #find the thrid quartile
    IQR = Q3 - Q1                #subtract to get IQR
    print(Q1)
    print(Q3)
    
    #filter out all valuse outside of IQR
    filter_df = df[(Q1 < df[col]) & (df[col] < Q3)]
    
    #return the specific column to check if the values are correct
    #return filter_df[[col]]
    #return the final dataframe
    return filter_df
            

remove(NHL_Goalies, 'GP')
```
